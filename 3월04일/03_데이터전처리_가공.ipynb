```python
# 데이터프레임 새로 만들기
# 데이터프레임 이름 : df_month_age
df_month_age=pd.DataFrame()
df_month_age[""]=df_20TABLE_org["수진자연령"]
df_month_age[""]=df_20TABLE_org["요양개시일자"]
df_month_age[""]=df_20TABLE_org["심결본인부담금"]
```

```python
df_month_age=df_20TABLE_org[["수진자연령","요양개시일자","심결본인부담금"]]
```

```python
# 9세미만, 10대, 20대, ~~ 90대, 100세이상
df_month_age
```

```python
# 9세이하 만들기
df_9under=df_month_age[(df_month_age["수진자연령"]<=9)]
df_9under.head(2)
# index 재배열
df_9under = df_9under.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "9세이하"넣기 ....
df_9under.head(2)
df_9under["연령대"]="9세이하"
df_9under.head(2)
```

```python
# 각 데이터프레임에 추가할 컬럼명 : 10대, 20대, ~~90대
# 데이터프레임 변수명 : df_10,~df_90
# 10대 만들기
df_10=df_month_age[(df_month_age["수진자연령"] >=10) & \
                    (df_month_age["수진자연령"] <=19)]
# index 재배열
df_10 = df_10.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "10대"넣기 ....
df_10["연령대"]="10대"
df_10.head(2)
```

```python
# 20대 만들기
df_20=df_month_age[(df_month_age["수진자연령"] >=20) & \
                    (df_month_age["수진자연령"] <=29)]
# index 재배열
df_20 = df_20.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "20대"넣기 ....
df_20["연령대"]="20대"
df_20.head(2)
```

```python
# 30대 만들기
df_30=df_month_age[(df_month_age["수진자연령"] >=30) & \
                    (df_month_age["수진자연령"] <=39)]
# index 재배열
df_30 = df_30.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "30대"넣기 ....
df_30["연령대"]="30대"
df_30.head(2)
```

```python
# 40대 만들기
df_40=df_month_age[(df_month_age["수진자연령"] >=40) & \
                    (df_month_age["수진자연령"] <=49)]
# index 재배열
df_40 = df_40.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "40대"넣기 ....
df_40["연령대"]="40대"
df_40.head(2)
```

```python
# 50대 만들기
df_50=df_month_age[(df_month_age["수진자연령"] >=50) & \
                    (df_month_age["수진자연령"] <=59)]
# index 재배열
df_50 = df_50.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "50대"넣기 ....
df_50["연령대"]="50대"
df_50.head(2)
```

```python
# 60대 만들기
df_60=df_month_age[(df_month_age["수진자연령"] >=60) & \
                    (df_month_age["수진자연령"] <=69)]
# index 재배열
df_60 = df_60.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "60대"넣기 ....
df_60["연령대"]="60대"
df_60.head(2)
```

```python
# 70대 만들기
df_70=df_month_age[(df_month_age["수진자연령"] >=70) & \
                    (df_month_age["수진자연령"] <=79)]
# index 재배열
df_70 = df_70.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "70대"넣기 ....
df_70["연령대"]="70대"
df_70.head(2)
```

```python
# 80대 만들기
df_80=df_month_age[(df_month_age["수진자연령"] >=80) & \
                    (df_month_age["수진자연령"] <=89)]
# index 재배열
df_80 = df_80.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "80대"넣기 ....
df_80["연령대"]="80대"
df_80.head(2)
```

```python
# 90대 만들기
df_90=df_month_age[(df_month_age["수진자연령"] >=90) & \
                    (df_month_age["수진자연령"] <=99)]
# index 재배열
df_90 = df_90.reset_index(drop=True)
#"연령대"컬럼 추가 후 값은 모두 "90대"넣기 ....
df_90["연령대"]="90대"
df_90.head(2)
```

```python
# df_month_age_any 데이터프레임 새로만들어서.
# 위에 df_9under~df_90까지 모두 행단위로 추가해 주세요..
# 모두 추가 후 인덱스 번호 재배열 까지 완료 후 조회..
# 데이터 프레임 연령대를 행 위아래로 합치기 순서는 언더9~ 90대
df_month_age_any=pd.concat([df_9under,df_10,df_20,df_30,
                            df_40,df_50,df_60,df_70,
                            df_80,df_90],ignore_index=True)
# index 재배열
df_month_age_any=df_month_age_any.reset_index(drop=True)
df_month_age_any
# 총 5,631건 확인 ....
```
