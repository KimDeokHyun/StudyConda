## 데이터 합하기

```python
# 시리즈 데이터 합하기
# 시리즈변수명 s1 ==> 값은 0, 1 / 인덱스 값은 A, B로 시리즈 만들기
# 시리즈변수명 s2 ==> 값은 2, 3, 4 / 인덱스 값은 a, b,c로 시리즈 만들기

s1 = pd.Series([0,1],
                index = ['A','B'])
s2 = pd.Series([2,3,4],
                index = ['a','b','c'])
```

```python
pd.concat([s1,s2], axis=0)
```

### 컬럼 단위로 합치기

```python
df_temp1 = pd.DataFrame([["Dog",3], ["Cat",10], ["Tiger",6], ["Mouse",3]],
                       index=[0,1,2,3],
                       columns = ["동물", "나이"])

df_temp1
```

```python
df_temp2 = pd.DataFrame([["집",0], ["초원",1], ["수풀",0], ["초원",1]],
                       index=[0,1,2,3],
                       columns = ["사는곳", "뿔 갯수"])

df_temp2
```

```python
#두개 결합
pd.concat([df_temp1, df_temp2], axis=1)
```
