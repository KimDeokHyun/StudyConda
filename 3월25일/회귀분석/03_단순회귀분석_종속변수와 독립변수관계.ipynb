```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager, rc
import platform

plt.rc('font', family = 'Malgun Gothic')
```

```python
df = pd.read_csv("./file_new/auto-mpg_v2.csv")
df
```

```python
# 지도학습 모델 : 단순회귀분석 (선형회귀모델, 선형회귀예측모델)
# 종속변수
# 직선 방정식 사용 : y = ax + b
# y : 종속변수
# x : 독립변수
# a = 기울기
# b = y절편

```

```python
#사용할 변수를 mpg, cylinders, horsepower, weight
ndf = df [["연비", "실린더", "엔진출력","중량"]]
ndf
```

```python
# 산점도 그래프 그리기
# 종속변수 : mpg, 독립변수 : weight
y = df["연비"]
x = df["중량"]
sns.regplot(x='중량', y='연비', ci=None, data=ndf, color='red')
plt.scatter(x, y, c = "RED",s = 10 )
plt.show()
```

```python
fig = plt.figure(figsize = (10, 8))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.regplot(x="중량", y = "연비", ax = ax1, color = 'red', fit_reg = True, data = df)
sns.regplot(x="중량", y = "연비", ax = ax2, color = 'blue', fit_reg = False, data = df)

plt.show()
plt.close()
```

```python
# jointplot 사용하여 회귀선이 있는것과 없는 것 그리기
sns.jointplot(x=ndf['연비'],y=ndf['중량'],color='green')
sns.jointplot(x=ndf['연비'],y=ndf['중량'],color='purple',kind='reg')
```
